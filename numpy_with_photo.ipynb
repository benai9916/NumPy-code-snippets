# Creating structured array

```python
import numpy as np
from skimage import io
```

```python
# covert image to nd array
photo = io.imread('demo.png')
pic_to_array = np.array(photo)
type(pic_to_array)
```

```python
pic_to_array.shape
```

## visualize photo using matplotlib 

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(pic_to_array)
```

```python
# slicing on photo
# start and stop is normal and step is -1 so we are reversing  the row and hence the image will be upside down
plt.imshow(pic_to_array[::-1])
```

```python
# all row are same, but we have reverse the column
plt.imshow(pic_to_array[:, ::-1])
```

```python
# just take the section of the photo
plt.imshow(pic_to_array[0:60, 0:80])
```

```python
# this will make the image size half
plt.imshow(pic_to_array[::2, ::2])
```

## applying mathematical function to numpy array

```python
# taking sin of element in the array
pic_to_array
photo_sin = np.sin(pic_to_array)
photo_sin
```

```python
# we can do some very efficine and quick statistic using nummpy arrays using various build in functions
# sum of all the element
print(np.sum(pic_to_array))
```

```python
# product of all the element
print(np.prod(pic_to_array))
```

```python
# standard deviation
print(np.std(pic_to_array))
```

```python
# variance
print(np.var(pic_to_array))
```

```python
# minimun
print(np.min(pic_to_array))
```

```python
# maximum
print(np.max(pic_to_array))
```

## comparison operation

```python
my_array =  np.arange(1,10)
my_array
```

```python
my_array < 5
```

```python
my_array > 4
```

```python
my_array == 4
```

```python
# get the elements as an array which are greater then 4
my_array[my_array > 4]
```

```python
# letc do this operation on image
# we took the threshold as 100 and anything greater then 100 will be replace with 255 and anything less then 100 will be replaced with 0
photo_masked  = np.where(pic_to_array > 100, 255, 0)
plt.imshow(photo_masked)
```

```python
# to get the transpose of an array that is to interchange the row and column
plt.imshow(pic_to_array[:,:,0].T)
```
